2.5. Измените типы данных в столбцах, где это необходимо, объясните
изменения.

```python
missing_values = ['not available', 'unknown', 'N/A', '-', 'NaN', 'nan']
df['budget'] = df['budget'].replace(missing_values, '0')
df['box_office'] = df['box_office'].replace(missing_values, '0')

# Преобразуем в float (все пустые строки станут 0)
df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0)
df['box_office'] = pd.to_numeric(df['box_office'], errors='coerce').fillna(0)
```

Преобразуем run_time в минуты (например, 2h 22m -> 142) и приведем к типу данных float.

```python
def convert_runtime_to_minutes(runtime):
    try:
        hours, minutes = 0, 0
        runtime = runtime.strip()
        if 'h' in runtime and 'm' in runtime:
            hours = float(runtime.split('h')[0].strip())
            minutes = float(runtime.split('h')[1].split('m')[0].strip())
        elif 'h' in runtime:
            hours = float(runtime.split('h')[0].strip())
        elif 'm' in runtime:
            minutes = float(runtime.split('m')[0].strip())
        return hours * 60 + minutes
    except:
        return np.nan

# Преобразуем столбец run_time в минуты
df['run_time'] = df['run_time'].apply(convert_runtime_to_minutes)

# Проверяем тип данных
df['run_time'] = df['run_time'].astype(float)
```

3.6. Используя функции и методы библиотеки Pandas (или иных библиотек)
проведите исследовательский анализ данных, способствующий решению
поставленной бизнес-задачи. Здесь нет конкретных инструкций и заданий –
поищите интересные закономерности в данных.

```python
# 1. Описательная статистика
print("Описательная статистика числовых столбцов:")
print(df[['rating', 'run_time', 'budget', 'box_office']].describe())

# 2. Анализ жанров: топ-5 жанров по количеству фильмов
top_genres = df['genre'].value_counts().head(5)
print("\nТоп-5 жанров по количеству фильмов:")
print(top_genres)

# 3. Корреляционный анализ
print("\nКорреляционная матрица:")
corr_matrix = df[['rating', 'run_time', 'budget', 'box_office', 'year']].corr()
print(corr_matrix)

# Визуализация корреляционной матрицы
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Корреляционная матрица')
plt.show()

# 4. Распределение рейтингов по жанрам (для топ-5 жанров)
top_genre_list = top_genres.index
df_top_genres = df[df['genre'].isin(top_genre_list)]
plt.figure(figsize=(10, 6))
sns.boxplot(x='genre', y='rating', data=df_top_genres)
plt.title('Распределение рейтингов по топ-5 жанрам')
plt.xticks(rotation=45)
plt.show()

# 5. Зависимость рейтинга от бюджета
plt.figure(figsize=(10, 6))
sns.scatterplot(x='budget', y='rating', data=df)
plt.title('Зависимость рейтинга от бюджета')
plt.xlabel('Бюджет ($)')
plt.ylabel('Рейтинг')
plt.show()

# 6. Группировка по годам (декадам) и анализ рейтингов
df['decade'] = (df['year'] // 10) * 10
decade_stats = df.groupby('decade').agg({
    'rating': ['mean', 'count'],
    'budget': 'mean',
    'box_office': 'mean'
}).round(2)
decade_stats.columns = ['rating_mean', 'movie_count', 'budget_mean', 'box_office_mean']
print("\nСтатистика по десятилетиям:")
print(decade_stats)

# Визуализация среднего рейтинга по десятилетиям
plt.figure(figsize=(10, 6))
sns.lineplot(x=decade_stats.index, y=decade_stats['rating_mean'])
plt.title('Средний рейтинг по десятилетиям')
plt.xlabel('Десятилетие')
plt.ylabel('Средний рейтинг')
plt.show()
```
